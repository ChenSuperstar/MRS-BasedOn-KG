```python
import os
import pandas as pd
import json 
```

```python
data_path = 'data/output.csv'

#节点
diseases = [] #疾病名称
departments = [] #疾病科室
symptoms = [] #疾病症状
drugs = [] #疾病用药
checks = [] #疾病检查
cures = [] #治疗方法
producers =[]  #品大类
foods = [] #食物


#实体关系
rel_department = [] #科室-科室关系
rel_category = [] #疾病-室关系
rel_symptom = [] # 疾病-症状关系
rel_acompany = [] #疾病并发关系
rel_commondrug = [] #疾病-通用药品关系
rel_recommanddrug = [] # 疾病-推荐药品关系
rel_check = [] #疾病-检查关系
rel_drug_producer = [] #厂商-药物关系
rel_cureway = [] # 疾病-治疗方式关系
rel_noteat = [] #疾病-忌吃
rel_doeat = [] #疾病-宜吃
rel_recommandeat = [] #疾病-推荐吃

```

```python
with open('data/medical.json',encoding='utf-8') as f:
    for line in f:
        disease_dic = {}
        obj = json.loads(line)
        disease = obj['name']
        diseases.append(disease)
        
        disease_dic['name'] = disease
        disease_dic['symptom'] = ''
        disease_dic['cure_department'] = ''
        disease_dic['cause'] = '' 
        disease_dic['get_prob'] = ''
        disease_dic['easy_get'] = ''        
        disease_dic['prevent'] = ''
        disease_dic['cure_way'] = ''
        disease_dic['curelasttime'] = ''
        disease_dic['cured_prob'] = ''
        
        if 'cure_department' in obj:
            cure_department = obj['cure_department']
            if len(cure_department) == 1:
                rel_category.append([disease, cure_department])
            if len(cure_department) == 2 :
                big = cure_department[0]
                small = cure_department[1]
                rel_department.append([small, big])
                rel_category.append([disease, small])
            departments += cure_department

        if 'symtom' in obj:
            symptom = obj['symptom']
            for symp in symptom:
                rel_symptom.append([disease, symptom])
            symptoms += obj['symptom']
        
        if 'common_drug' in obj:
            common_drug = obj['common_drug']
            for drug in common_drug:
                rel_commondrug.append([disease, drug])
            drugs += common_drug
        
        if 'recommand_drug' in obj:
            recommand_drug = obj['recommand_drug']
            for drug in recommand_drug:
                rel_recommanddrug.append([disease, drug])
            drugs += recommand_drug
        
        if 'check' in obj:
            check = obj['check']
            for che in check:
                rel_check.append([disease,check])
        
        if 'cure_way' in obj:
            cure_way = obj['cure_way']
            for cure in cure_way:
                rel_cureway.append([disease, cure])
            cures += cure_way
        
        if 'drug_detail' in obj:
            drug_detail = obj['drug_detail']
            producer = [i.split('(')[0] for i in drug_detail]
            rel_drug_producer += [[i.split('(')[0], i.split('(')[-1].replace(')', '')] for i in drug_detail]
            producers += producer        
        
            
        
        
        
        
        
```

```python
data_path = 'data/output.csv'

diseases = []
symptoms = []
drugs = []

rel_symptom = []
rel_drug = []
```

```python
data = pd.read_csv(data_path)
data
```

```python
for index, row in data.iterrows():
    disease_dic = {}
    disease = row['name']
    disease_dic['name'] = disease
    disease_dic['drug'] = ''
    disease_dic['symptom'] = ''
    diseases.append(disease)
    
    
    
    if len(eval(row['symptom'])) > 0:
        symptoms += eval(row['symptom'])
        for symptom in eval(row['symptom']):
            rel_symptom.append([disease, symptom])
    
        
    if len(eval(row['drug'])) > 0:
        drugs += eval(row['drug'])
        for drug in eval(row['drug']):
            rel_drug.append([disease, drug])
```

```python
disease_list = list(set(diseases))
symptom_list = list(set(symptoms))
drug_list = list(set(drugs))
disease_symptom_list = rel_symptom
disease_drug_list = rel_drug
```

```python
disease_symptom_list
```

```python
# 写入entity2id.txt文件
with open('data/entity2id.txt', 'w') as f:
    # 写入实体总数
    f.write(str(len(disease_list) + len(symptom_list) + len(drug_list)) + '\n')
    # 写入disease的id
    for i, disease in enumerate(disease_list):
        f.write(disease + '\t' + str(i) + '\n')
    # 写入symptom的id
    for i, symptom in enumerate(symptom_list):
        f.write(symptom + '\t' + str(len(disease_list) + i) + '\n')
    # 写入drug的id
    for i, drug in enumerate(drug_list):
        f.write(drug + '\t' + str(len(disease_list) + len(symptom_list) + i) + '\n')


```

```python
# 写入relation2id.txt文件
with open('data/relation2id.txt', 'w') as f:
    # 写入关系总数
    f.write('2\n')
    # 写入disease和symptom之间的关系
    f.write('disease_symptom\t0\n')
    # 写入disease和drug之间的关系
    f.write('disease_drug\t1\n')
```

```python
with open('data/train2id.txt', 'w') as f:
    # 写入三元组总数
    f.write(str(len(disease_symptom_list) + len(disease_drug_list)) + '\n')
    # 写入disease和symptom之间的三元组
    for disease, symptom in disease_symptom_list:
        f.write(str(disease_list.index(disease)) + '\t' + str(len(disease_list) + symptom_list.index(symptom)) + '\t0\n')
    # 写入disease和drug之间的三元组
    for disease, drug in disease_drug_list:
        f.write(str(disease_list.index(disease)) + '\t' + str(len(disease_list) + len(symptom_list) + drug_list.index(drug)) + '\t1\n')
```

```python
symptom_drug_dict = {}

# 遍历疾病-症状列表
for disease, symptom in disease_symptom_list:
    # 如果症状已经在字典中，则直接加入对应药物
    if symptom in symptom_drug_dict:
        symptom_drug_dict[symptom] += [disease_drug[1] for disease_drug in disease_drug_list if disease_drug[0] == disease]
    # 如果症状不在字典中，则创建新的症状-药物关系
    else:
        symptom_drug_dict[symptom] = [disease_drug[1] for disease_drug in disease_drug_list if disease_drug[0] == disease]

# 打印症状-药物字典
print(symptom_drug_dict)
```

```python
symptom_drug_dict['紫绀']
```

```python
symptom_drug_data = []

# 遍历字典中的每个键值对
for symptom, drugs in symptom_drug_dict.items():
    # 将每个药品和症状构成一个元组，并添加到列表中
    for drug in drugs:
        symptom_drug_data.append((symptom, drug))
```

```python
symptom_drug_df = pd.DataFrame(symptom_drug_data, columns=['Symptom', 'Drug'])
symptom_drug_df.to_csv('data/symptom_drug_dict.csv', index=False)
```
